## Section 4: Step Activation Function

**What this section is about:**

## Section 5: Simple Perceptron Prediction

**What this section is about:**  
This section combines dot product, bias, and activation into a single prediction function.  
This is a complete single-neuron model.

image.png

```python
def predict(x, w, b):
    z = np.dot(x, w) + b
    return step_function(z)

x = np.array([2, 1])
w = np.array([1.0, -0.5])
b = -0.2

print("Prediction:", predict(x, w, b))
```

### ✏️ Exercise 5
Change the values of `w` and `b`.  
Observe how the prediction changes.

```python
def step_function(z):
    return 1 if z >= 0 else 0

def predict(x, w, b):
    z = np.dot(x, w) + b
    return step_function(z)

x = np.array([2, 1])

w = np.array([1.0, -0.5])
b = -0.2
print("Original:", predict(x, w, b))

w = np.array([0.5, 0.5])
b = -1.0
print("Test 1:", predict(x, w, b))

w = np.array([-1.0, 2.0])
b = 0.5
print("Test 2:", predict(x, w, b))

w = np.array([0.0, 0.0])
b = -0.1
print("Test 3:", predict(x, w, b))
```

## Section 6: Data Visualization with Matplotlib

**What this section is about:**  
Visualization helps us understand how data points are distributed.  
This section plots 2D data used for perceptron training.

```python
import matplotlib.pyplot as plt

X = np.array([[2,3], [1,1], [2,1], [3,2]])
y = np.array([1, 0, 0, 1])

plt.scatter(X[:,0], X[:,1], c=y)
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.title("Training Data")
plt.show()
```

### ✏️ Exercise 6

## Section 7: Decision Boundary Preview

**What this section is about:**  
A perceptron learns a straight-line decision boundary.  
This section previews how weights and bias define that line.

```python
w = np.array([1, -1])
b = 0

x_vals = np.linspace(0, 4, 100)
y_vals = -(w[0] * x_vals + b) / w[1]

plt.scatter(X[:,0], X[:,1], c=y)
plt.plot(x_vals, y_vals)
plt.title("Decision Boundary Preview")
plt.show()
```

### ✏️ Exercise 7
Change the values of `w` and `b` and observe how the line moves.

```python
import numpy as np
import matplotlib.pyplot as plt

X = np.array([[2,3], [1,1], [2,1], [3,2]])
y = np.array([1, 0, 0, 1])

fig, axes = plt.subplots(1, 3, figsize=(15, 4))

params = [
    (np.array([1, -1]), 0),
    (np.array([2, -1]), -1),
    (np.array([0.5, -1]), 1),
]

x_vals = np.linspace(0, 4, 100)

for ax, (w, b) in zip(axes, params):
    y_vals = -(w[0] * x_vals + b) / w[1]
    ax.scatter(X[:,0], X[:,1], c=y, cmap='coolwarm', s=100, edgecolors='black')
    ax.plot(x_vals, y_vals, 'k--', linewidth=2)
    ax.set_title(f"w={w}, b={b}")
    ax.set_xlim(0, 4)
    ax.set_ylim(0, 4)

plt.tight_layout()
plt.show()
```
